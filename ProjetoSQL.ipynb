```python
#Encontre o número de livros lançados depois de 1 de janeiro de 2000.
query1 = """
SELECT 
    COUNT(DISTINCT book_id) as total_livros
FROM
    books
WHERE
    publication_date > '2000-01-01';
"""

resultado1 = pd.read_sql(query1, engine)

print("=== TAREFA 1 ===")
print("Número de livros lançados depois de 1º de janeiro de 2000:")
print(resultado1)
```

```python
#Encontre o número de avaliações e a classificação média para cada livro.
query2 = """
SELECT 
    r.book_id,
    b.title,
    COUNT(r.rating) as total_avaliacoes,
    AVG(r.rating) as classificacao_media
FROM ratings r
JOIN books b ON r.book_id = b.book_id
GROUP BY r.book_id, b.title
ORDER BY total_avaliacoes DESC;
"""

resultado2 = pd.read_sql(query2, engine)

print("=== TAREFA 2 ===")
print("Número de avaliações e classificação média para cada livro:")
print(resultado2)
```

#### Comentário:
O total das avaliações variam bastante, indo de 1 até 160 por exemplo, já a classificação média se mantém na média de 3 a 4,5, sendo maior que isso é um livro de destaque e menor um livro não tão aclamado pelo público. Temos como Top1 de avaliações o livro 'Twilight' (crepúsculo).

```python
#Identifique a editora que lançou o maior número de livros com mais de 50 páginas 
query3 = """
SELECT
    p.publisher, 
    COUNT(*) as total_livros
FROM
    books b
JOIN publishers p ON b.publisher_id = p.publisher_id
WHERE b.num_pages > 50
GROUP BY p.publisher
ORDER BY total_livros DESC
LIMIT 1;
"""

resultado3 = pd.read_sql(query3, engine)

print("=== TAREFA 3 ===")
print("Editora que lançou o maior número de livros com mais de 50 páginas:")
print(resultado3)
```

#### Comentário: 
A editora 'Penguin Books' foi a editora que lançou mais livros com mais de 50 páginas!

```python
#Identifique o autor com a média mais alta de classificação de livros: olhe apenas para livros com pelo menos 50 classificações.
query4 = """
WITH book_ratings AS (
    SELECT
        b.book_id,
        b.author_id,
        AVG(r.rating) AS avg_rating,
        COUNT(r.rating_id) AS num_ratings
    FROM books AS b
    JOIN ratings AS r ON b.book_id = r.book_id
    GROUP BY b.book_id, b.author_id
    HAVING COUNT(r.rating_id) >= 50
)
    

SELECT
    br.author_id,
    a.author,
    AVG(br.avg_rating) AS author_avg_rating
FROM book_ratings AS br
JOIN authors AS a ON br.author_id = a.author_id
GROUP BY br.author_id, a.author
ORDER BY author_avg_rating DESC
LIMIT 1;

"""
resultado4 = pd.read_sql(query4, engine)
print("=== TAREFA 4 ===")
print("Autor com a média mais alta de classificação de livros (mínimo 50 classificações):")
print(resultado4)
```

#### Comentário:
A autora J.K Rowling, famosa pela saga de Harry Potter, é a que tem a média mais alta de classificações de livros, qual será o gênero que ela mais escreve? Seria uma métrica boa para avaliarmos. Junto com ela temos a Mary GrandPré, será ela uma co-escritora ou parceira? 

```python
#Encontre o número médio de avaliações entre usuários que classificaram mais do que 50 livros.
query5 = """
SELECT
    AVG(review_count) AS media_livros_classificados
FROM (
    SELECT 
        COUNT(DISTINCT book_id) AS review_count
    FROM 
        reviews
    WHERE 
        username IN(
            SELECT
                username
            FROM
                ratings
            GROUP BY 
                username
            HAVING
                COUNT(DISTINCT book_id) > 50
        )
    GROUP BY username
) AS reviews_counts;

"""
resultado5 = pd.read_sql(query5, engine)
print("=== TAREFA 5 ===")
print("Número médio de avaliações entre usuários que classificaram mais do que 50 livros:")
print(resultado5.round(2))
```

A média de avaliações entre os usuários que classificaram mais do que 50 livros é de: 24 avaliações
